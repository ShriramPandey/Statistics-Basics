#**#1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

**Descriptive Statistics** summarize and describe the main features of a dataset. They provide a snapshot of the data, allowing us to understand its basic characteristics. Examples include:

*   **Measures of Central Tendency:** Mean, median, and mode (e.g., the average height of students in a class).
*   **Measures of Dispersion:** Range, variance, and standard deviation (e.g., the spread of test scores).
*   **Frequency Distributions:** Tables or graphs showing how often each value occurs (e.g., a bar chart showing the number of people in different age groups).

**Inferential Statistics** use data from a sample to make inferences or predictions about a larger population. They allow us to draw conclusions beyond the data we have directly observed, often involving probability. Examples include:

*   **Hypothesis Testing:** Determining if there is enough evidence to support a claim about a population (e.g., testing if a new drug is more effective than an old one).
*   **Confidence Intervals:** Estimating a range of values that is likely to contain a population parameter (e.g., estimating the average income of a city's residents with a certain level of confidence).
*   **Regression Analysis:** Examining the relationship between variables to make predictions (e.g., predicting a student's test score based on the number of hours they studied).

In simple terms, descriptive statistics describe the data you have, while inferential statistics help you make educated guesses about data you don't have.

# **#2. What is sampling in statistics? Explain the differences between random and stratified sampling.**
- **Sampling** in statistics is the process of selecting a subset of individuals or items from a larger population to collect data and make inferences about the entire population. It's often impractical or impossible to collect data from every member of a population, so sampling allows us to gather representative data efficiently.

There are various sampling methods, each with its own advantages and disadvantages. Two common types are random sampling and stratified sampling:

*   **Random Sampling (Simple Random Sampling):**
    *   In simple random sampling, every member of the population has an equal and independent chance of being selected for the sample.
    *   This method helps minimize bias and ensures that the sample is likely representative of the population.
    *   Think of it like drawing names out of a hat – each name has the same chance of being picked.
    *   **Example:** To survey student opinions in a school, you could assign a number to each student and use a random number generator to select a sample of students.

*   **Stratified Sampling:**
    *   In stratified sampling, the population is divided into subgroups (strata) based on shared characteristics (e.g., age, gender, income level).
    *   Then, a random sample is taken from *each* stratum.
    *   This method is useful when you want to ensure that certain subgroups are adequately represented in your sample, especially if those subgroups are small relative to the overall population.
    *   **Example:** If you are surveying voters and want to ensure representation from different age groups, you could divide the population into age strata (e.g., 18-25, 26-40, 41-60, 60+) and then randomly select a certain number of voters from each stratum.

**Key Differences:**

*   **Random Sampling:** Selects individuals from the entire population randomly.
*   **Stratified Sampling:** Divides the population into subgroups and then randomly selects individuals from each subgroup.

Stratified sampling can be more complex to implement than simple random sampling but can provide a more representative sample when dealing with diverse populations and the need to ensure representation of specific subgroups.

# **#3. Define mean, median, and mode. Explain why these measures of central tendency are important.**

**Mean, median, and mode** are measures of central tendency, which are single values that attempt to describe a set of data by identifying the central position within that set of data.

*   **Mean (Average):**
    *   Calculated by summing all the values in a dataset and dividing by the number of values.
    *   Sensitive to outliers (extreme values).
    *   Represents the "balancing point" of the dataset.
    *   **Example:** The mean of the numbers 2, 4, 6, 8 is (2+4+6+8)/4 = 5.

*   **Median:**
    *   The middle value in a dataset that is ordered from least to greatest.
    *   If the dataset has an even number of values, the median is the average of the two middle values.
    *   Not affected by outliers.
    *   Represents the "middle" value of the dataset.
    *   **Example:** The median of the numbers 2, 4, 6, 8, 10 is 6. The median of 2, 4, 6, 8 is (4+6)/2 = 5.

*   **Mode:**
    *   The value that appears most frequently in a dataset.
    *   A dataset can have one mode (unimodal), multiple modes (multimodal), or no mode.
    *   Useful for categorical data.
    *   Represents the most common value in the dataset.
    *   **Example:** The mode of the numbers 2, 4, 4, 6, 8 is 4.

### Importance of Measures of Central Tendency

These measures are important because they provide a concise summary of the typical value in a dataset. They help us:

*   **Understand the center of the data:** Get a single value that represents the "average" or "typical" data point.
*   **Compare different datasets:** Easily compare the central values of two or more groups.
*   **Identify patterns and trends:** Observe how the central value changes over time or across different conditions.
*   **Inform decision-making:** Use the central value as a basis for making informed choices or drawing conclusions.

Choosing the appropriate measure of central tendency depends on the type of data and the goal of the analysis. For example, the median is often preferred over the mean when dealing with skewed data or outliers because it is less affected by extreme values.

# **4. Explain skewness and kurtosis. What does a positive skew imply about the data?**

**Skewness** is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it tells us the degree to which a distribution is skewed or lopsided to the left or right.

*   **Positive Skew (Right-Skewed):** The tail of the distribution is longer on the right side. This means there are more values concentrated on the left side of the distribution, with a few larger values stretching the tail to the right.
    *   **Implication:** A positive skew implies that the mean is typically greater than the median and the mode (if a single mode exists). This is because the few large values in the right tail pull the mean towards the right.
    *   **Example:** Income distribution in many populations is positively skewed, with most people earning a moderate income and a smaller number of people earning very high incomes.

*   **Negative Skew (Left-Skewed):** The tail of the distribution is longer on the left side. This means there are more values concentrated on the right side of the distribution, with a few smaller values stretching the tail to the left.
    *   **Implication:** A negative skew implies that the mean is typically less than the median and the mode. The few small values in the left tail pull the mean towards the left.
    *   **Example:** The distribution of exam scores when the exam was very easy might be negatively skewed, with most students scoring high and a few students scoring lower.

*   **Zero Skew:** The distribution is symmetrical. The mean, median, and mode are typically equal (in a perfectly symmetrical unimodal distribution).
    *   **Example:** A perfect normal distribution has zero skewness.

**Kurtosis** is a measure of the "tailedness" of the probability distribution of a real-valued random variable. It describes the shape of the tails and the peakedness of the distribution relative to a normal distribution.

*   **Leptokurtic (High Kurtosis):** The distribution has heavier tails and a sharper peak than a normal distribution. This indicates a higher probability of extreme values (outliers).
*   **Mesokurtic (Medium Kurtosis):** The distribution has kurtosis similar to a normal distribution.
*   **Platykurtic (Low Kurtosis):** The distribution has lighter tails and a flatter peak than a normal distribution. This indicates a lower probability of extreme values.


# **# 5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.**
**numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

```python
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Compute mean
mean_value = statistics.mean(numbers)
print(f"Mean: {mean_value}")

# Compute median
median_value = statistics.median(numbers)
print(f"Median: {median_value}")

# Compute mode
try:
    mode_value = statistics.mode(numbers)
    print(f"Mode: {mode_value}")
except statistics.StatisticsError as e:
    print(f"Could not compute mode: {e}")
```

# **#6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:**
- **list_x = [10, 20, 30, 40, 50]**
- **list_y = [15, 25, 35, 45, 60]**

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Compute covariance
covariance = np.cov(list_x, list_y)[0, 1]
print(f"Covariance: {covariance}")

# Compute correlation coefficient
correlation_coefficient = np.corrcoef(list_x, list_y)[0, 1]
print(f"Correlation Coefficient: {correlation_coefficient}")
```

# **# 7. Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:**
### **data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Calculate quartiles and IQR to identify outliers
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Define bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"First Quartile (Q1): {q1}")
print(f"Third Quartile (Q3): {q3}")
print(f"Interquartile Range (IQR): {iqr}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

# **#8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.**
##  - **Explain how you would use covariance and correlation to explore this relationship.**
##  - **Write Python code to compute the correlation between the two lists:**
##  * **advertising_spend = [200, 250, 300, 400, 500]**
##  * **daily_sales = [2200, 2450, 2750, 3200, 4000]**



**1. Covariance:**

*   **What it tells us:** Covariance measures the extent to which two variables change together. A positive covariance indicates that as one variable increases, the other tends to increase as well. A negative covariance indicates that as one variable increases, the other tends to decrease. A covariance close to zero suggests a weak linear relationship.
*   **How to use it:** I would calculate the covariance between the `advertising_spend` and `daily_sales` datasets. A positive covariance would suggest that higher advertising spend is associated with higher daily sales.
*   **Limitation:** The magnitude of covariance is dependent on the units of the variables, making it difficult to compare the strength of the relationship between different pairs of variables.

**2. Correlation:**

*   **What it tells us:** Correlation, specifically the Pearson correlation coefficient (r), is a standardized measure of the linear relationship between two variables. It ranges from -1 to +1.
    *   +1 indicates a perfect positive linear relationship.
    *   -1 indicates a perfect negative linear relationship.
    *   0 indicates no linear relationship.
*   **How to use it:** I would calculate the correlation coefficient between `advertising_spend` and `daily_sales`. The value of the correlation coefficient would tell us the strength and direction of the linear relationship. A value close to +1 would indicate a strong positive linear relationship, meaning increased advertising spend is strongly associated with increased daily sales. A value closer to 0 would suggest a weak or no linear relationship.
*   **Advantage over Covariance:** Correlation is unitless, making it easier to interpret and compare the strength of relationships across different datasets.

**In summary:**

I would first calculate the covariance to see if there is a general tendency for advertising spend and daily sales to move together. Then, I would calculate the correlation coefficient to quantify the strength and direction of the linear relationship, providing a more interpretable measure of how closely the two variables are related. A strong positive correlation would provide evidence to the marketing team that increasing advertising spend is likely to lead to an increase in daily sales.

## Python Code to Compute Correlation**

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute correlation coefficient
correlation_coefficient = np.corrcoef(advertising_spend, daily_sales)[0, 1]
print(f"Correlation Coefficient: {correlation_coefficient}")
```

# **9. Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use to understand the distribution of the following customer satisfaction survey data and write Python code to create a histogram using Matplotlib for the survey data:**
### **survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

# 1. Mean:
- The mean of the survey scores would represent the average customer satisfaction level.
- A higher mean would indicate a generally higher level of satisfaction among customers,
- while a lower mean would suggest lower overall satisfaction.
- It gives a single value that summarizes the central tendency of the data.

# 2. Standard Deviation:
- The standard deviation would measure the dispersion or spread of the satisfaction scores around the mean.
- A low standard deviation would indicate that the satisfaction scores are clustered closely around the mean,
- suggesting consistent satisfaction levels among customers.
- A high standard deviation would indicate that the scores are more spread out,
- implying greater variability in customer satisfaction (some very satisfied, some very dissatisfied).

# 3. Histogram:
- A histogram would visually display the frequency distribution of the survey scores.
- The x-axis would represent the satisfaction scores (e.g., from 1 to 10),
- and the y-axis would represent the frequency or count of how many customers gave each score or a score within a specific range (bin).
- The height of each bar would show how common each satisfaction level is.
- A histogram would help to:
    * See the shape of the distribution (e.g., is it skewed, is it bimodal?).
    * Identify the most frequent satisfaction scores (the mode).
    * Understand the range of scores and whether there are any outliers.
    * Get a visual sense of the spread and central tendency of the data.

```python
## Generate histogram code
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=7, edgecolor='black') # Set bins to cover the range of scores
plt.title("Distribution of Customer Satisfaction Survey Scores")
plt.xlabel("Satisfaction Score")
plt.ylabel("Frequency")
plt.show()
```

# Based on the generated histogram:

**1. General shape of the distribution:**
- The distribution appears to be roughly symmetrical, with a slight tendency towards higher scores.
- There isn't a strong skew to either the left or right.

**2. Most frequent score(s) or range of scores:**
- Looking at the bar heights, the scores 7, 8, and 9 appear to be the most frequent,
- with score 7 having the highest frequency.

**3. Spread of the scores and outliers:**
- The scores range from 4 to 10, indicating a moderate spread.
- The histogram doesn't visually suggest the presence of significant outliers,
- as most scores are clustered between 5 and 9, and the bars on the ends (4 and 10)
- are not extremely isolated or small compared to the others.
